# WEEKLY STOCK DATA

```python
import requests
import alpha_vantage
import pandas as pd
import csv
```

```python
API_URL = "https://www.alphavantage.co/query"
```

```python
stocks=["GOOG"]
```

```python
weekly_open_price=[]
weekly_close_price=[]
weekly_high_price=[]
weekly_low_price=[]
weekly_volume=[]
stock_name=[]
weekly_date=[]


for stock in stocks:
    data_weekly = {
        "function": "TIME_SERIES_WEEKLY",
        "symbol": stock,
        "outputsize": "full",
        "datatype": "json",
        "apikey":"Q9U4GSEBHNGWOD9R" }

    response_weekly = requests.get(API_URL, data_weekly)
    stock_data_weekly=response_weekly.json()
    values_weekly = stock_data_weekly["Weekly Time Series"]
    
    keys_weekly = (values_weekly.keys())

    for key in keys_weekly:
        stock_name.append(stock)
        weekly_date.append(key)
        weekly_open_price.append(values_weekly[key]['1. open'])
        weekly_close_price.append(values_weekly[key]['4. close'])
        weekly_high_price.append(values_weekly[key]['2. high'])
        weekly_low_price.append(values_weekly[key]['3. low'])
        weekly_volume.append(values_weekly[key]['5. volume'])
        
    
    
       
   
    
    
```

```python
weekly_stock_df=pd.DataFrame(stock_name, columns=["Stock"])
weekly_stock_df["Date"]= weekly_date
weekly_stock_df["Open"]=weekly_open_price
weekly_stock_df["Close"]=weekly_close_price
weekly_stock_df["High"]=weekly_high_price
weekly_stock_df["Low"]=weekly_low_price
weekly_stock_df["Volume"]=weekly_volume
weekly_stock_df
```

```python
weekly_stock_df.count()
```

```python
weekly_stock_df.to_excel("Output/alpha_vantage_data_weekly.xlsx", index=False)
```

# DAILY STOCK DATA

```python
daily_open_price=[]
daily_close_price=[]
daily_high_price=[]
daily_low_price=[]
daily_volume=[]
stock_name=[]
daily_date=[]


for stock in stocks:
    data_daily = {
        "function": "TIME_SERIES_DAILY",
        "symbol": stock,
        "outputsize": "full",
        "datatype": "json",
        "apikey":"Q9U4GSEBHNGWOD9R" }

    response_daily = requests.get(API_URL, data_daily)
    stock_data_daily=response_daily.json()
    

    values_daily = stock_data_daily["Time Series (Daily)"]

    
    keys_daily = (values_daily.keys())
    for key in keys_daily:
        stock_name.append(stock)
        daily_date.append(key)
        daily_open_price.append(values_daily[key]['1. open'])
        daily_close_price.append(values_daily[key]['4. close'])
        daily_high_price.append(values_daily[key]['2. high'])
        daily_low_price.append(values_daily[key]['3. low'])
        daily_volume.append(values_daily[key]['5. volume'])
        
```

```python
daily_stock_df=pd.DataFrame(stock_name, columns=["Stock"])
daily_stock_df["Date"]=daily_date
daily_stock_df["Open"]=daily_open_price
daily_stock_df["Close"]=daily_close_price
daily_stock_df["High"]=daily_high_price
daily_stock_df["Low"]=daily_low_price
daily_stock_df["Volume"]=daily_volume
daily_stock_df
```

```python
daily_stock_df.to_excel("Output/alpha_vantage_data_daily.xlsx", index=False)
```
