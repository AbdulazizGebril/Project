```python
import numpy as np 
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("apple_stock_summary.csv") #loading the csv file
df.head()
```

```python
df_2 = df[df['Date'] >'6/15/2018']
df_2
```

```python
df_2=df_2.drop(["Stock","Date"], axis=1)
df_2.head()
```

```python
print('Number of rows before removing rows with missing values: ' + str(df_2.shape[0]))

# Replace ? with np.NaN
df_2.replace('?', np.NaN, inplace=True)

# Remove rows with np.NaN
df_2.dropna(how='any', inplace=True)

print('Number of rows after removing rows with missing values: ' + str(df_2.shape[0]))
```

```python
X = df_2.drop("Percent_change", axis=1).values

# Get the target vector
y = df_2["Percent_change"].values
```

```python
target = 'Percent_change'

# Get the target vector
y = df_2[target]
# Specify the name of the features
features = list(df_2.drop(target, axis=1).columns)

# Get the feature vector
X = df_2[features]
```

```python
#import numpy as np
# Labels are the values we want to predict
#labels = np.array(df_2['Percent_Change'])
# Remove the labels from the features
# axis 1 refers to the columns
#features= df_2.drop('Percent_Change', axis = 1)
# Saving feature names for later use
#feature_list = list(features.columns)
# Convert to numpy array
#features = np.array(features)
```

```python
#ONE HOT ENCODING
X = pd.get_dummies(X)

# Show the encoded features
X.head()
```

```python

from sklearn.model_selection import train_test_split

train_features, test_features, train_labels, test_labels = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
# Train the model on training data
rf.fit(train_features, train_labels);
```

```python
# Use the forest's predict method on the test data
predictions = rf.predict(test_features)
# Calculate the absolute errors
errors = abs(predictions - test_labels)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), )
predictions
errors
```

```python
#Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / test_labels)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
accuracy
mape
```

```python
# Get numerical feature importances
importances = list(rf.feature_importances_)
# List of tuples with variable and importance
feature_importances = [(feature, round(importance, 2)) for feature, importance in zip(features, importances)]
# Sort the feature importances by most important first
feature_importances = sorted(feature_importances, key = lambda x: x[1], reverse = True)
# Print out the feature and importances 
[print('Variable: {:20} Importance: {}'.format(*pair)) for pair in feature_importances];
```
